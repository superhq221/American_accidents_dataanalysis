```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Accident = pd.read_csv('data/Accident_US_2019.csv')
```

```python
Accident.columns
```

```python
import matplotlib.pyplot as plt
import numpy as np

Accident.plot(kind = 'scatter', x = 'Severity', y = 'Temperature(F)')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x = 'Severity', y= 'Temperature(F)', data = Accident);
```

```python
sns.jointplot(x = 'Severity', y = 'Temperature(F)', data = Accident)
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Severity', y='Temperature(F)', data=Accident)
plt.title('Temperature(F) & Severity', fontsize=24)
plt.xlabel('Severity', fontsize=12)
plt.ylabel('Temperature(F)', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Severity', y='Pressure(in)', data=Accident)
plt.title('Pressure(in) & Severity', fontsize = 24)
plt.xlabel('Severity', fontsize=12)
plt.ylabel('Pressure(in)', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Severity', y='Humidity(%)', data=Accident)
plt.title('Humidity(%) & Severity')
plt.xlabel('Severity', fontsize=12)
plt.ylabel('Humidity', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='Severity', y='Wind_Chill(F)', data=Accident)
plt.title('Wind Chill & Severity')
plt.xlabel('Severity', fontsize=12)
plt.ylabel('Wind_Chill(F)', fontsize=12)
plt.show()
```

## 交通指示标识和天气因素对车祸严重程度的相关程度探究

### 1.	概况
#### a)	严重程度分布
![1-a](1a.png)
 
-	2级事故占近 3/4（687，573），3级事故约占1/4。
-	只有不到3%的事故对交通造成非常严重的影响。

#### b)	各州城市的严重车祸数量
![1-b](1b.png)
 
-	2019 年大多数城市4级交通事故的数量都在 100 起以下。
-	然而，在佛罗里达州（FL）的几个城市的数字远远超过100，迈阿米是该国4级事故最多的城市（779）。

#### c)	各季节的不同严重程度车祸的数量
![1-c](1c.png)
 
-	在冬季交通事故最频繁，尤其是2级严重事故（220，501）。
-	然而在第一季发生的事故稍严重些7，4级事故605起，3级事故65，158起。

#### d)	各州车祸数量与平均严重程度
![1-d1](1d1.png)
![1-d2](1d2.png)
 
-	加州（CA）的事故数量（231，307）比其他州要高得多，接下来6个州的车祸数量也是4万到8万之间的巨大数字，但幸运的是，其严重程度并不高。
-	三个事故发生最少（低于200）的州均位于美国中北部地区，平均事故严重程度超过3.0。
-	也许在交通比较发达的地方，车祸数量会增加，但同时对车祸的影响也会得到更快速的解决；反之，地理环境对发展交通不友好，车流量会少，恢复交通能力弱。

### 2.	天气因素
![2-a](2a.png)
 
-	事故大多发生在天气晴朗时，事故总数超过30万起。
-	事故发生在雪、雷、雷、雷暴等恶劣天气中，平均严重程度高于正常天气（如晴天或多云天）发生的事故。
-	3级事故的当天平均降水量明显较高。

### 3.	交通因素
#### a)	白天或夜晚 × 有无交通指示
![3-a](3a.png)
 
-	白天发生在交通指示附近的事故的平均严重程度最低。
-	夜间没有交通指示的路段上平均交通意外严重程度最高。
-	交通部门应在高速公路上设置更多的交通标志，帮助司机预测路段，减少交通事故的发生。

#### b)	不同交通指示附近的事故严重程度
![3-b](3b.png)
 
-	大多数交通事故附近没有交通标志，平均严重程度达到2.352。
-	junction交通指示附近的交通事故导致的影响最为严重（2.428）。

### 4.	发现与建议
a)	佛罗里达州要更注重减少重大交通事故数量，尤其对于几个4级事故高发城市。  
b)	美国中西部的几个州尽管车祸数量少，但要重视车祸发生时带来的影响。  
c)	在恶劣天气中发生的事故更严重，降水量或许能影响交通事故的严重程度。  
d)	政府应该在马路上设置更多的交通信号灯或标志，这样很可能减低交通事故严重程度，附近有路况指示的交通事故平均严重性低于没有指示的道路上的严重程度。  
e)	在晚上开车需要更加集中注意力。
